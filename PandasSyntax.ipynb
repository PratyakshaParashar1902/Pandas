```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, np.nan, np.nan],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df.head()
```

```python
# Dropping columns which have na values
df.dropna(axis = 1)
```

```python
# Dropping rows if all columns have na values
df.dropna(how = 'all', axis =1)
```

```python
dupli = pd.DataFrame({"name": ['Alfred', 'Alfred', 'Catwoman'],
                   "toy": ['Alfred', 'Batman', 'Catwoman'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
dupli
```

```python
# Duplicate function - it does not take axis as parameter but do take subset  
dupli.drop_duplicates(subset='name')
```

```python
# Keeps last duplicate by default it keeps first one
dupli.drop_duplicates(subset='name', keep='last')
```

```python
# Does not keep any of the duplicates
dupli.drop_duplicates(subset='name', keep=False)
```
